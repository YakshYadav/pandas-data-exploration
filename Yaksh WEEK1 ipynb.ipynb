In this notebook i will try to explore and clean a shopping dataset using pandas.

```python
# importing required libraries
import pandas as pd
import numpy as np

print("libraries imported")
```

## Step 1 - Load Dataset

```python
df = pd.read_csv('shopping_dataset.csv')
print("dataset loaded!")
print("rows:", df.shape[0])
print("columns:", df.shape[1])
```

## Step 3 - Handle Missing Values

```python
# checking how many null values are there
df.isnull().sum()
```

```python
# lets also check percentage of missing values
missing_pct = df.isnull().sum() / len(df) * 100
print(missing_pct)
```

```python
# filling missing values with median 
# using median because it is less affected by outliers compared to mean
df['age'].fillna(df['age'].median(), inplace=True)
df['price'].fillna(df['price'].median(), inplace=True)
df['review_rating'].fillna(df['review_rating'].median(), inplace = True)

print("done filling missing values")

# verify
print("remaining nulls:", df.isnull().sum().sum())
```

## Step 4 - Filtering Rows and Selecting Columns

```python
# filter - age between 25 to 45 and payment method is credit card
filtered_df = df[(df['age'] >= 25) & (df['age'] <= 45) & (df['payment_method'] == 'Credit Card')]
print("filtered rows:", len(filtered_df))
filtered_df.head()
```

```python
# selecting only some useful columns
selected_cols = df[['customer_id', 'age', 'gender', 'category', 'item_purchased', 'price', 'quantity', 'payment_method']]
selected_cols.head()
```

```python
# how many purchases in each category
df['category'].value_counts()
```

```python
df['payment_method'].value_counts()
```

## Step 5 - Remove Duplicates

```python
print("before removing duplicates:", len(df))

dups = df.duplicated().sum()
print("duplicate rows:", dups)

df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

print("after removing duplicates:", len(df))
```

## Step 6 - Creating New Column: total_amount

total_amount = price * quantity

```python
df['total_amount'] = df['price'] * df['quantity']
df['total_amount'] = df['total_amount'].round(2)

print("total_amount column added")
print("min:", df['total_amount'].min())
print("max:", df['total_amount'].max())
print("mean:", df['total_amount'].mean().round(2))
```

```python
df[['customer_id','item_purchased','price','quantity','total_amount']].head(8)
```

```python
# revenue by category
rev_by_cat = df.groupby('category')['total_amount'].sum().sort_values(ascending=False)
print(rev_by_cat)
```

## Step 7 - Save the Cleaned Dataset

```python
df.to_csv('shopping_dataset_cleaned.csv', index=False)
print("cleaned file saved!")
print("final shape:", df.shape)
```
